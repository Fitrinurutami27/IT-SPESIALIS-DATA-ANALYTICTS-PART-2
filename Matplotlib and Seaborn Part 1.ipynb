## Using Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95] #ton per hectare
```

```python
plt.plot(yield_apples)
```

```python
years = [2017, 2018, 2019, 2020, 2021, 2022]
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95]
```

```python
plt.plot(years, yield_apples) #plt.plot(x, y)
```

```python
#To make it easier and clearer, it can be changed

plt.plot(years, yield_apples)
plt.xlabel('Years')
plt.ylabel('Yield(ton per hectare)');
```

```python
apples = [0.89, 0.91, 0.93, 0.97, 0.986, 0.943]
tangerines = [0.79, 0.61, 0.93, 0.97, 0.962, 0.931]
```

```python
# comparison with 2 data

plt.plot(years, apples)
plt.plot(years, tangerines)
plt.xlabel('Years')
plt.ylabel('Yield(ton per hectare)');
```

```python
# giving names to lines, dots and giving titles

plt.plot(years, apples, marker='o')
plt.plot(years, tangerines, marker='x')

plt.xlabel('Years')
plt.ylabel('Yield(ton per hectare)')

plt.title("Crop yield in kanto")
plt.legend(['Apples', 'tangerines'])
```

```python
# How to make the resulting visualization image enlarge

plt.figure(figsize=(12,6))

plt.plot(years, apples, marker='o')
plt.plot(years, tangerines, marker='x')

plt.xlabel('Years')
plt.ylabel('Yield(ton per hectare)')

plt.title("Crop yield in kanto")
plt.legend(['Apples', 'tangerines'])
```

```python
import pandas as pd
```

```python
crop_yield = dict(tahun=years, apel=apples, jeruk=tangerines)
data_panen = pd.DataFrame(crop_yield)
```

```python
data_panen
```

```python
ax = sns.lineplot(x='tahun', y='apel', data= data_panen)
ax1 = sns.lineplot(x= 'tahun', y='jeruk', data = data_panen)

plt.show()
```
