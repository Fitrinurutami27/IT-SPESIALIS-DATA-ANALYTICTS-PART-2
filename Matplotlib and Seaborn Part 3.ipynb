```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris_df = sns.load_dataset("iris")
iris_df.head(3)
```

```python
import numpy as np
```

```python
plt.hist(iris_df.sepal_width, bins=np.arange(2, 5, 0.25));
```

```python
plt.hist(iris_df.sepal_width, bins=[1,3,4,4,5])
```

```python
sentosa_df = iris_df[iris_df.species == 'sentosa']
versicolor_df = iris_df[iris_df.species == 'versicolor']
virginica_df = iris_df[iris_df.species == 'virginica']
```

```python
plt.hist(sentosa_df.sepal_width, alpha = 0.4)
plt.hist(versicolor_df.sepal_width, alpha = 0.4)
```

```python
plt.title('Distribution of sepal width')

plt.hist([sentosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
        stacked=True)
plt.legend(['sentosa', 'vesicolor', 'virginica'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Pengunjung_Mall.csv")
```

```python
df.head()
```

```python
# Using DataFrame 'df' to create a barplot
sns.barplot(x='Gender', y='Pengeluaran_USD', data=df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

pengeluaran = df['Pengeluaran_USD']

# NumPy to manually define bin limits
bin_edges = np.histogram_bin_edges(pengeluaran, bins= 5)

# histogram
plt.hist(pengeluaran, bins=bin_edges, edgecolor='k')
plt.xlabel('Pengeluaran (USD)')
plt.ylabel('Frekuensi')
plt.title('Histogram Pengeluaran Menggunakan NumPy')
plt.show()
```

```python
# Using a list of bin restrictions
custom_bins = [0, 20, 40, 60, 80, 100]

# Create a histogram with custom bin boundaries
plt.hist(pengeluaran, bins=custom_bins, edgecolor='k')
plt.xlabel('Pengeluaran (USD)')
plt.ylabel('Frekuensi')
plt.title('Histogram Pengeluaran dengan Batasan Bin Kustom')
plt.show()
```

```python
# histogram for each group of data
plt.hist(df[df['Gender'] == 'Pria']['Pengeluaran_USD'], bins=9, edgecolor='k', alpha=0.5, label='Pria')
plt.hist(df[df['Gender'] == 'Wanita']['Pengeluaran_USD'], bins=9, edgecolor='k', alpha=0.5, label='Wanita')

plt.xlabel('Pengeluaran (USD)')
plt.ylabel('Frekuensi')
plt.title('Histogram Pengeluaran untuk Pria dan Wanita')
plt.legend()
plt.show()
```
