# using Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
years = [2017, 2018, 2019, 2020, 2021, 2022]
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95]
```

```python
# To make it easier and clearer, it can be changed

plt.plot(years, yield_apples)
plt.xlabel('Years')
plt.ylabel('Yield(ton per hectare)');
```

```python
apples = [0.89, 0.91, 0.93, 0.97, 0.986, 0.943]
tangerines = [0.79, 0.61, 0.93, 0.97, 0.962, 0.931]
```

```python
# comparison with 2 data

plt.plot(years, apples)
plt.plot(years, tangerines)
plt.xlabel('Years')
plt.ylabel('Yield(ton per hectare)');
```

```python
plt.bar(years, tangerines) #plt.bar(x, y)

plt.xlabel('Years')
plt.ylabel('Yield(ton per hectare)')

plt.title("crop yield in kanto")
```

```python
plt.bar(years, apples)
plt.bar(years, tangerines, bottom=apples)
```

# Task- kamis2

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Pengunjung_Mall.csv")
```

```python
df.head()
```

```python
# Menggunakan DataFrame 'df' untuk membuat barplot
sns.barplot(x='Gender', y='Pengeluaran_USD', data=df)
```

```python
plt.hist(df['Pengeluaran_USD'], bins=10, edgecolor='k');
```

```python
plt.hist(df['Pengeluaran_USD'], bins=5, edgecolor='k');
```

```python
import matplotlib.pyplot as plt
import numpy as np

pengeluaran = df['Pengeluaran_USD']

# NumPy untuk menentukan batasan bins secara otomatis
bin_edges = np.histogram_bin_edges(pengeluaran, bins= 5)

# histogram
plt.hist(pengeluaran, bins=bin_edges, edgecolor='k')
plt.xlabel('Pengeluaran (USD)')
plt.ylabel('Frekuensi')
plt.title('Histogram Pengeluaran Menggunakan NumPy')
plt.show()
```
